```python
#Data Types

# 1. Numeric Data Types : In Python we have 3 Numeric data types - int, float and complex
# int data type : int is used to represent the whole numbers Example -10,0,5,7
# let us define a variable of int data type
positive_num = 10
negative_num = -100
zero_num = 0
print("---- Int Data Type ----")
print(positive_num)
print(negative_num)
print(zero_num)
print("_______________________________",end="\n\n")

# Float Data type: Float data type is used to represent the decimal values ((example: 10.5, -3.2, 0.99))
pdeci_number = 10.55
ndeci_number = -5.33
print("---- float Data Type ----")
print(pdeci_number,ndeci_number,sep=",")
print("_______________________________",end="\n\n")

# Complex data type : Numbers with real and imaginary parts (example: 3+5j)
com_num = 3+4j
real_num = 4+0j
imaginary_num = 0+5j
print("---- Complex Data Type ----")
print(com_num,real_num,imaginary_num, sep = ",")
print("_______________________________",end="\n\n")
```

```python
#Sequential data types : Used to store multiple items in a single variable.
# We have
#--------> String
#--------> List
#--------> Tuple
# List :Whenever we want to store data of the same or different types where the insertion order is preserved, duplicates are allowed,
#and the data can be changed (mutable), we use a List.
#Python lists are heterogeneous, meaning they can hold multiple types of objects in one container.
print("------------------------------ List ------------------------------")

# Same Data Type
numbers = [1, 2, 3, 4, 5]
print(numbers)   # [1, 2, 3, 4, 5]

# Mixed Data Type
mixed = [1, "Venu", 3.14, True, None]
print(mixed)  # [1, 'Venu', 3.14, True, None]

# we can also store nested data inside list (Nested data - List inside a list)
nested = [1, [2, 3], ["apple", "banana"], (10, 20)]
print(nested) # [1, [2, 3], ['apple', 'banana'], (10, 20)]

#We can also store list of dictionaries
students = [ {"name": "Venu", "age": 25},{"name": "Kiran", "age": 24}]
print(students[0]["name"])   # Venu

#------------------------------ Tuple ------------------------------
print("------------------------------ Tuple ------------------------------")
# Tuple : Whenever we want to store data of the same or different types where the insertion order is preserved, duplicates are allowed, but the data cannot be changed (immutable) after creation — we use a Tuple.

# Tuple definition
num_tuple = (1,2,3,4,5,6)
print(num_tuple)

# Tuple with different data types
mixed_tuple = (1,"Venu",[1,2,3],{"city":"Hartford"})
print(mixed_tuple)

#Accessing elements in tuple using index
print(mixed_tuple[3]["city"])

#Duplicates are allowed
duplicate_tuple = (1,2,3,4,5,4,3,2,1,7)
print(duplicate_tuple)

#Nested Tuple
nested = ((1, 2), (3, 4), (5, 6))
print(nested[1])       # (3, 4)

#Example 7: Tuple with One Element. When a tuple has only one element, you must add a comma after it.
single = (10,)
print(type(single))

single_not_tuple = (10)
print(type(single_not_tuple))

#Tuple is immutable
fruits = ("Apple","Carrot")
fruits(1,"tata")
```

```python
#Mapping Data Type: Whenever we needed to store the data in a Key-Value pair we have to use this Mapping Data Types, Dictionaries comes under Mapping Data Types.

# Dictionary : Whenever we want to store data in the form of key–value pairs, where each key is unique, the insertion order is preserved (in Python 3.7+), and the data is changeable (mutable) — we use a Dictionary.
#Creating a Dictionary
student_data = {"name":"Venu Madhav Pendurthi","age":29,"city":"Hartford"}
print(student_data)

#Accessing values using keys
print(student_data["name"])
print(student_data.get("age"))

#Adding or Updating Items
student_data["city"] = "Boston"
print(student_data)

#Removing Items

#a. Remove single element
student_data.pop("age")
print(student_data)

#b. Remove all data from the dictionary
student_data.clear()
print(student_data)

#c. Delete the entire dictionary
#del student_data
#print(student_data)

#Iterating Through a Dictionary
student = {"name": "Kiran", "age": 24, "grade": "A"}

for item in student:
  print(item, ":", student[item])

#Dictionary Methods
student = {"name": "Kiran", "age": 24, "grade": "A"}
print(student.keys())
print(student.values())
print(student.items())
```

```python
#Random Data Types : Sets
# Set: Whenever we want to store multiple unique items where the insertion order is not important, duplicates are not allowed, and the data is changeable (mutable) — we use a Set.

num_set = {1,2,3,4}
print(num_set)

num_set_duplicate = {1,2,3,2,1,4}
print(num_set_duplicate) # even tough duplicates are inserted when printing we will get only unique values

num_set_mixed = {1,"venu",True,3.14}
print(num_set_mixed)

#Adding relements to Set
string_set = {"Venu","Kesav","Teja"}
string_set.add("Hari")
print(string_set)

#Removing elements from set
string_set.remove("Venu")
print(string_set)

#Accessing elements from set
for item in string_set:
  print(item)

print(string_set[1]) # since set is an un ordered collection of data, it doesn't supports indexis so we cannot rtrive data using indexing.
```
